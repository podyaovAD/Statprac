```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt('populations.txt', dtype ={'names': ('year', 'hare', 'lynx', 'carrot'), 
                                  'formats': ('i4', 'f4', 'f4', 'f4')})
```

```python
plt.plot(data['year'], data['hare'])
plt.plot(data['year'], data['lynx'])
plt.plot(data['year'], data['carrot'])
plt.legend(('hare', 'lynx', 'carrot'))
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
mas=data['hare']/(data['lynx']+data['hare'])*100
plt.bar(data['year'], mas)
plt.title('Hare')
plt.xlabel('years')
plt.ylabel('percents')
plt.show()
plt.bar(data['year'],100-mas)
plt.title('Lynx')
plt.xlabel('years')
plt.ylabel('percents')
plt.show()
```

```python
print('mean and std')
print('Hare:   {}, {}'.format(np.mean(data['hare']), np.std(data['hare'])))
print('Lynx:   {}, {}'.format(np.mean(data['lynx']), np.std(data['lynx'])))
print('Carrot: {}, {}'.format(np.mean(data['carrot']), np.std(data['carrot'])))
plt.hist(data['hare'], 10)
plt.title('Hare')
plt.axvline(x=np.mean(data['hare']),color='green')
plt.show()
plt.hist(data['lynx'], 10)
plt.title('Lynx')
plt.axvline(x=np.mean(data['lynx']),color='green')
plt.show()
plt.hist(data['carrot'], 10)
plt.title('Carrot')
plt.axvline(x=np.mean(data['carrot']),color='green')
plt.show()
```

```python
print('Max years')
print('Lynx:   {}'.format(data['year'][data['lynx'].argmax()]))
print('Hare:   {}'.format(data['year'][data['hare'].argmax()]))
print('Carrot: {}'.format(data['year'][data['carrot'].argmax()]))
plt.plot(data['year'],data['hare'],'r')
plt.plot(data['year'],data['lynx'],'g')
plt.plot(data['year'],data['carrot'],'b')
plt.legend(('Hare','Lynx','Carrot'))
plt.xlabel('years')
plt.ylabel('population')
plt.axvline(x=data['year'][data['hare'].argmax()],color='r',linestyle='--')
plt.axvline(x=data['year'][data['lynx'].argmax()],color='g',linestyle='--')
plt.axvline(x=data['year'][data['carrot'].argmax()],color='b',linestyle='--')
plt.show()
```

```python
mas1 = np.sign(data['hare']-data['lynx'])
mas2 = list(map(lambda x: 'H' if x == 1 else 'L', mas1))
plt.pie([mas2.count('H'), mas2.count('L')], labels=['H', 'L'], autopct='%1.f')
plt.show()
```

```python
print ('population>50000')
print('Hare:   {}'.format(data['year'][np.where(data['hare']>50000)]))
print('Lynx:   {}'.format(data['year'][np.where(data['lynx']>50000)]))
print('Carrot: {}'.format(data['year'][np.where(data['carrot']>50000)]))
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(('hare','lynx','carrot'))
plt.axis([1900,1920,0,80000])
plt.plot(data['year'],data['hare'],'r')
plt.plot(data['year'],data['lynx'],'g')
plt.plot(data['year'],data['carrot'],'b')
plt.vlines(data['year'][np.where(data['hare']>50000)],0,80000,color='r',linestyle='--',alpha=0.5)
plt.vlines(data['year'][np.where(data['lynx']>50000)],0,80000,color='g',linestyle='--',alpha=0.5)
plt.vlines(data['year'][np.where(data['carrot']>50000)],0,80000,color='b',linestyle='--',alpha=0.5)
plt.show()
```

```python
print('2 years lowest populations')
print('Hare:   {}'.format(data['year'][np.argsort(data['hare'])[0:2]]))
print('Lynx:   {}'.format(data['year'][np.argsort(data['lynx'])[0:2]]))
print('Carrot: {}'.format(data['year'][np.argsort(data['carrot'])[0:2]]))
```
